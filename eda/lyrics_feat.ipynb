```python
import pandas as pd
from nltk.corpus import stopwords
from textblob import TextBlob
```

```python
lyrics = pd.read_csv('data/lyrics.csv')
billboard = pd.read_csv('data/billboard.csv')
```

```python
billboard.head()
```

```python
relevant = billboard[billboard.date >= '2000-01-01']
```

```python
relevant.head()
```

```python
lyrics.head()
```

```python
df = pd.merge(relevant, lyrics, on=['artist', 'song'])
```

```python
date_mean = df.groupby(['date'])['sentiment'].mean()
date_subj = df.groupby(['date'])['subjectivity'].mean()
```

```python
df_merged = pd.merge(date_mean, date_subj, on=["date"])
```

```python
df_merged.to_csv('data/lyrics_sent_subj.csv')
```
