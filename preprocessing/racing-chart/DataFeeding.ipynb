```python
import numpy as np
import pandas as pd
```

```python
their = pd.read_csv("share_rank.csv").drop(columns= "rank").drop([1, 11, 111, 569])
their.year = their.year.apply(lambda x: str(x)+".1")
their.lastValue = their.value
their.to_csv("new_share_rank.csv")
```

```python
our_df = pd.read_csv("our_rank.csv")
our_df
```

```python
year_artist = our_df.groupby(['year', "artist"]).count()['rank'].reset_index().sort_values(["year", "rank"], ascending = False).groupby("year").head(15).reset_index(drop=True) # get count for each year, no duplicate for year-artist pair now.
our_df = our_df.drop_duplicates(['year','artist'])
our_df1 = our_df[['artist', 'weeks-on-board', 'year']].merge(year_artist[["year", "artist"]])
```

```python
# our_df1["lastValue"] = our_df1[["year", "weeks-on-board"]].groupby(['year']).shift()

our_df1["lastValue"] = our_df1['value']
```

```python
our_df1 = our_df1.rename(columns = {"weeks-on-board":"value", "artist":"name"})
```

```python
our_df1 = our_df1.replace(np.NaN, 0)
our_df1
```

```python
our_df1.to_csv("our_df1.csv")
```
